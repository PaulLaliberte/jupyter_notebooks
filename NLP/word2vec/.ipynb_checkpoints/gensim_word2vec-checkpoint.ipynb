# Experimentation with Gensim word2vec

```python
import gensim

import pandas as pd
```

```python
df = pd.read_pickle('../class/hw3/data/more_reviews.pkl')
df
```

```python
sentences = [i.split() for i in df['sentence'].values]
```

```python
model = gensim.models.Word2Vec(sentences, size=100, window=5, min_count=4)
```

```python
model.wv['hotel']
```

```python
model.wv.most_similar(positive=['clean'], negative=['gross'])
```

```python
model.wv.similarity('friendly', 'pleasant')
```

```python
model.wv.doesnt_match('quiet pleasant gross nice'.split())
```
