```python
import pandas as pd
import numpy as np

from tqdm import tqdm_notebook

from guess_language import guess_language

from mtranslate import translate
```

```python
df = pd.read_pickle('author_affiliations.pkl')
df
```

```python
affils = (df['Affiliations'].fillna(' ')).values
greatest_affil = []

for i in tqdm_notebook(affils):
    s = i[0].split(',')[0]
    if s == ' ':
        greatest_affil.append(np.nan)
    else:
        language = guess_language(s)
        if language == 'en':
            greatest_affil.append(s)
        else:
            try:
                greatest_affil.append(translate(s, 'en', language))
            except Exception as e:
                greatest_affil.append(s)
```

```python
lower_affil = []
for i in greatest_affil:
    if type(i) == float:
        lower_affil.append(' ')
    else:
        lower_affil.append(i.lower())

len(lower_affil)
```

```python
keywords = ['school', 'schools', 'university', 'college', '(mit)']
academic_labels = []

found = False
for affil in lower_affil:
    s = affil.split()
    for key in keywords:
        if key in s and found == False:
            academic_labels.append(1)
            found = True
            break
            
    if found == False:
        academic_labels.append(0)
        
    found = False
        
        

        
len(academic_labels)
```

```python
df.drop('Affiliations', axis=1, inplace=True)
df['Affiliations'] = lower_affil
df['Academic (1)'] = academic_labels
df
```

```python
df.to_pickle('academics_nonacademics.pkl')
df.to_csv('academics_nonacademics.csv', encoding='utf-8-sig')
```
