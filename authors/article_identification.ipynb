```python
import glob

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt

from tqdm import tqdm_notebook
```

```python
workingpapers_english = pd.read_csv('../model_workingpapers/full_english/english_workingpapers_final.csv', encoding='utf-8-sig')
workingpapers_english.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
workingpapers_english = workingpapers_english.rename(columns={'4' : 'JEL Code'})
```

```python
workingpapers_english
```

```python
journals_english = pd.read_csv('../model_journals/full_english/english_journals_final.csv')
journals_english.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
journals_english = journals_english.rename(columns={'3' : 'JEL Code'})
```

```python
journals_english
```

```python
authors_df = pd.read_pickle('academics_nonacademics.pkl')
authors_df
```

```python
authors = authors_df['Name'].values
academic = authors_df['Academic (1)'].values

academic_authors = [sorted(authors[i].split()) for i in range(len(authors)) if academic[i] == 1]
academic_authors
```

```python
wp_new = workingpapers_english
wp_new.drop(['0', '1', '2', '3', 'Combined', 'JEL Code'], axis=1, inplace=True)
wp_new
```

```python
j_new = journals_english
j_new.drop(['0', '1', '2', 'Combined', 'JEL Code'], axis=1, inplace=True)
j_new
```

```python
number_indices_wp = [i for i in range(0, 1111145)]
number_indices_j = [i for i in range(0, 1412131)]
```

```python
not_string = False
academia_rows_wp = []

for i in tqdm_notebook(number_indices_wp):
    row = wp_new.iloc[i]
    for j in row:
        if type(j) == str:
            s = sorted(j.replace(',', '').strip().split())
            if len(s) > 1:
                for j in academic_authors:
                    if s == j:
                        academia_rows_wp.append(i)
                        break
        else:
            not_string = True
            break
            
    if not_string == True:
        continue      
```

```python
not_string = False
academia_rows_j = []

for i in tqdm_notebook(number_indices_j):
    row = j_new.iloc[i]
    for j in row:
        if type(j) == str:
            s = sorted(j.replace(',', '').strip().split())
            if len(s) > 1:
                for j in academic_authors:
                    if s == j:
                        academia_rows_j.append(i)
                        break
        else:
            not_string = True
            break
            
    if not_string == True:
        continue

        
```

```python
academia_df = pd.DataFrame()
nonacademia_df = pd.DataFrame()

for i in tqdm_notebook(number_indices_wp[:100]):
    if i in academia_rows_wp:
        academia_df = academia_df.append(workingpapers_english.iloc[i])
    else:
        nonacademia_df = nonacademia_df.append(workingpapers_english.iloc[i])
        
for i in tqdm_notebook(number_indices_j[:100]):
    if i in academia_rows_j:
        academia_df = academia_df.append(journals_english.iloc[i])
    else:
        nonacademia_df = nonacademia_df.append(journals_english.iloc[i])
```

```python
academia_df.drop('4', axis=1, inplace=True)
academia_df.to_pickle('dataframe_pickles/academia.pkl')
academia_df.to_csv('dataframe_pickles/academia.csv', encoding='utf-8-sig')
academia_df
```

```python
nonacademia_df.drop('4', axis=1, inplace=True)
nonacademia_df.to_pickle('dataframe_pickles/nonacademia.pkl')
nonacademia_df.to_csv('dataframe_pickles/nonacademia.csv', encoding='utf-8-sig')
nonacademia_df
```

```python
counts_academia = {'D' : 0, 'E' : 0, 'F' : 0, 'G' : 0, 'H' : 0, 'I' : 0, 'J' : 0, 'K' : 0,
          'L' : 0, 'O' : 0}

counts_nonacademia = {'D' : 0, 'E' : 0, 'F' : 0, 'G' : 0, 'H' : 0, 'I' : 0, 'J' : 0, 'K' : 0,
          'L' : 0, 'O' : 0}

academia_jels = academia_df['JEL Code'].values
nonacademia_jels = nonacademia_df['JEL Code'].values
```

```python
for i in tqdm_notebook(academia_jels):
    if type(i) != float:
        jels = i.split()
        for j in jels:
            if j in counts_academia:
                counts_academia[j] += 1
```

```python
counts_academia
```

```python
for i in tqdm_notebook(nonacademia_jels):
    if type(i) != float:
        jels = i.split()
        for j in jels:
            if j in counts_nonacademia:
                counts_nonacademia[j] += 1
```

```python
counts_nonacademia
```

>Plot our results.

```python
#ACADEMIA

plt.figure()

plt.bar(range(len(counts_academia)), counts_academia.values(), align='center')
plt.xticks(range(len(counts_academia)), counts_academia.keys())
plt.title('Academic: JEL Code Frequency')

plt.savefig('academia.png')

plt.show()
```

```python
#NON-ACADEMIA
plt.figure()

plt.bar(range(len(counts_nonacademia)), counts_nonacademia.values(), align='center')
plt.xticks(range(len(counts_nonacademia)), counts_nonacademia.keys())
plt.title('Non-Academic: JEL Code Frequency')

plt.savefig('nonacademia.png')

plt.show()
```

```python
#plot adapted from: https://stackoverflow.com/questions/28931224/adding-value-labels-on-a-matplotlib-bar-chart

data = [(i,j) for i,j in counts_academia.items()]
data = sorted(data, key=lambda x : x[0])

frequencies = [i[1] for i in data]
x_labels = [i[0] for i in data]

freq_series = pd.Series.from_array(frequencies) 

# now to plot the figure...
plt.figure(figsize=(12, 8))
ax = freq_series.plot(kind='bar')
ax.set_title("Academic: JEL Code Frequency")
ax.set_xlabel("JEL Code")
ax.set_ylabel("Frequency")
ax.set_xticklabels(x_labels, rotation=0)

rects = ax.patches

# Now make some labels
bar_labels = frequencies

for rect, label in zip(rects, bar_labels):
    height = rect.get_height()
    ax.text(rect.get_x() + rect.get_width() / 2, height + 1, label, ha='center', va='bottom')

plt.savefig("image.png")
```
