```python
import gensim
import pyLDAvis

import pyLDAvis.gensim as gensimvis
```

```python
dictionary = gensim.corpora.dictionary.Dictionary.load('topic_model_data/merged_dict_t13.pkl')
corpus = gensim.corpora.MmCorpus('topic_model_data/merged_corpus_t13.mm')
model = gensim.models.LdaMulticore.load('topic_model_data/lda_bow_t13.model')
```

```python
vis_data = gensimvis.prepare(model, corpus, dictionary, mds='mmds', n_jobs=-1)
```

```python
pyLDAvis.display(vis_data)
```
