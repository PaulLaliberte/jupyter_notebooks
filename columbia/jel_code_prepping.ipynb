# Task 1: Binary Labeling of JEL Codes
***

We prep for fitting our model by labeling entries with 'F' (and soon to be 'L') codes and remove digits in the JEL Codes.

```python
import re
import glob
import string


import pandas as pd
import numpy as np
```

```python
#Get a dataframe of all csv's combined
counter = 0
df = None
for filename in glob.glob('papers_csv_grid/*.csv'):
    if counter == 0:
        #pass russian journal for now
        counter += 1
        pass
    elif counter == 1:
        df = pd.read_csv(filename, header=None, low_memory=False)
        counter += 1
    elif counter < 1000:
        new_df = pd.read_csv(filename, header=None, low_memory=False)
        new_df = new_df.drop([0])
        df = pd.concat([df, new_df])
        counter += 1
    else:
        break
        
df.to_csv('dataframe_pickles/original_frame_1000.csv', encoding='utf-8-sig')      
df
```

```python
df_jel = df[[2,4]]
df_jel = df_jel.dropna()
df_jel
```

```python
#label jel codes
#find root codes
arr = df_jel[4].values[1:]

found = False
labels = ['F']
root_codes = ['Root Codes']

tmp = []
for a in arr:
    for b in a.split():
        for c in b:
            if c.isdigit() == False and c not in string.punctuation:
                tmp.append(c.upper())
                if c == 'F':
                    labels.append(1)
                    found = True
                    break
        
        if found == True:
            break
            
    root_codes.append(list(set(tmp)))
    tmp =[]
    if found == False:
        labels.append(0)
        
    found = False
    
df_jel.drop(4, axis=1, inplace=True)
df_jel
```

```python
df_jel[4] = np.array(root_codes, dtype=object)
df_jel[5] = np.array(labels)
df_jel
```

```python
#This saves the dataframe as a 'pickle.' 
df_jel.to_csv('dataframe_pickles/f_codes_extended_1000.csv', encoding='utf-8-sig')
```
